### 03 -  猜數字（Guess the number）
#### 請使用者猜一個介於範圍（例如 1 到 100）之間的數字，
#### 若使用者猜錯，就詢問他們想繼續玩還是退出；若使用者猜對，就顯示祝賀訊息，
#### 並統計使用者的嘗試次數。如果使用者輸入的數字超出設定範圍，就顯示錯誤提示。

![](data\guess.png)


````python
import random
Ans = random.randint(0, 101)
MAX = 100
MIN = 0
count = 0

while True:
    print("猜數字範圍 :", MIN, "< ? <", MAX)
    num = int(input("請輸入0~100的數字 : "))
    count += 1
    if num == Ans:
        print("Bingo!!", "你猜了", count, "次")
        break
    elif num > Ans:
        MAX = num
        print("再小一點!", "你猜了", count, "次")
    else:
        MIN = num
        print("再大一點!", "你猜了", count, "次")
````

### 04 - 計算字數（Word count）
#### 使用者輸入一段文字或讀取一檔案，程式統計字數。

> 輸入(Input):我要成為寫程式的專家 <br>
輸出(Output):你用了 10 個文字述說內心的想法

#### 在特定的文章字串中搜尋輸入的特定字，請使用者輸入欲搜尋的文字，印出總共有幾個。<br>
>輸入(Input):天<br>
輸出(Output): 9

* 進階:計算文章中去除標點符號後的字數，找出文章中出現最多的字與次數

````python
word = input()
print("你用了 %d 個文字述說內心的想法"%(len(word)))


s = '''庭院深深深幾許，楊柳堆煙，簾幕無重數。玉勒雕鞍遊冶處，樓高不見章臺路。
雨橫風狂三月暮，門掩黃昏，無計留春住。淚眼問花花不語，亂紅飛過鞦韆去。
'''
# 輸入依自算字數出現次數
word = input("請輸入一個字 :")
count = 0
for i in s:
    if i == word:
        count += 1
print(word, "出現了", count, "次")

# 將輸入的文字去除標點符號
punctuation = ["'", "。", "，", "\n", "；", "：", "？", "「", "」", "'"]
for j in s:
    if j in punctuation:
        s = s.replace(j, "")
print(s)

# 計算出現最多次的字
maxchar = 0
maxword = 0
for k in s:
    times = s.count(k)
    if times > maxchar:
        maxword = k
        maxchar = times
print(maxword, "是出現最多的字", maxchar, "次")
````

### 06 - 及格名單（Pass list）
#### 請使用集合功能來完成各科級名單的判斷
```
米花市帝丹小學一年级B班正舉辦期中考試
數學及格的有：柯南、灰原、步美、美環、光彦
英文及格的有：柯南、灰原、丸尾、野口、步美
以上已列出全班所有人
```

>請分別列出<br>
數學及格但英文不及格的同學名單<br>
數學不及格但英文及格的同學名單<br>
兩者皆及格名單<br>
* Hint:差集(減法)、交集


````python
math = {'柯南', '灰原', '步美', '美環', '光彦'}
eng = {'柯南', '灰原', '丸尾', '野口', '步美'}

# 數學及格但英文不及格的同學名單
x = set(math) - set(eng)
print(x)
# 數學不及格但英文及格的同學名單
y = set(eng) - set(math)
print(y)
# 兩者皆及格名單
z = math & eng
print(z)
````
